# What city sold the most product?

```python
retaildata.head()
```

```python
def get_city(address):
    return address.split(",")[1].strip(" ")

def get_state(address):
    return address.split(",")[2].split(" ")[1]

retaildata['City'] = retaildata['Purchase Address'].apply(lambda x: f"{get_city(x)}  ({get_state(x)})")
retaildata.head()
```

```python
retaildata.groupby(['City']).sum()['Total_price']
```

```python
import matplotlib.pyplot as plt

keys = [city for city, df in retaildata.groupby(['City'])]

plt.bar(keys,retaildata.groupby(['City']).sum()['Total_price'])
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')
plt.xticks(keys, rotation='vertical', size=8)
plt.show()
```

# What products are most often sold together?Â¶

```python
retaildata
```

```python
df=retaildata[retaildata['Order ID'].duplicated(keep=False)]
df.head(20)
```

```python
df['Grouped']=df.groupby('Order ID')['Product'].transform(lambda x:','.join(x))
df.head(20)
```

```python
df=df[['Order ID','Grouped']].drop_duplicates()
df.head(10)
```

```python
from itertools import combinations
from collections import Counter

count =Counter()

for row in df['Grouped']:
    row_list =row.split(',')
    count.update(Counter(combinations(row_list, 3)))
    
count.most_common(10)
```
