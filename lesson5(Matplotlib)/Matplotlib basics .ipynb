# Simple graph

```python
from  matplotlib import pyplot as plt 

# X,y data 
x=[1,2,3]
y=[4,5,6]

#plot canvas
plt.plot(x,y)

#show graph 
plt.show
```

# ADD title ,lables and ploting variables

```python
from  matplotlib import pyplot as plt 

# X,y data 
x=[1,2,3]
y=[4,5,6]

#plot canvas
plt.plot(x,y)


#title 
plt.title('Simple graph ')
plt.ylabel('Y axis ')
plt.xlabel('x axis')
#show graph 
plt.show
```

# Adding style to graph 

```python
from  matplotlib import pyplot as plt 
from matplotlib import style 

style.use('ggplot')

# X,y data 
x=[5,8,10]
y=[12,16,6]

# X2,y2 data 
x2=[6,9,12]
y2=[6,15,7]

#plot canvas
plt.plot(x,y,'g',label='one',linewidth=7)
plt.plot(x2,y2,'c',label='two',linewidth=7)

plt.legend()
#title 
plt.title('Simple graph ')
plt.ylabel('Y axis ')
plt.xlabel('x axis')

#https://matplotlib.org/stable/gallery/color/named_colors.html
plt.grid(True,color='k')
#show graph 
plt.show
```
